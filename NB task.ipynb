```python
df=df.drop_duplicates()
```

```python
from collections import Counter
import re
def get_words(text):
    text=text.lower()
    text=re.sub(r"[^a-zA-Z\s]","",text)
    words=text.split()
    return words
all_words=df[" Review"].apply(get_words).sum()
word_counts=Counter(all_words)
print(word_counts.most_common(10))
```

```python
all_words
```

```python
df["tokens"]=df[" Review"].str.split()
```

```python
df["token_len"]=df["tokens"].apply(len)
```

```python
df.head()
```

```python
from nltk.corpus import stopwords
```

```python
stop_words=set(stopwords.words("english"))
```

```python
df['tokens_no_stopwords']=df['tokens'].apply(lambda tokens:[word for word in tokens if word not in stop_words])
```

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
df['tokens_stemmed']=df['tokens_no_stopwords'].apply(lambda tokens:[stemmer.stem(word) for word in tokens])
```

```python
df.head()
```

```python
df["cleaned"]=df["tokens_stemmed"].apply(lambda x:" ".join(x))
```

```python
df["cleaned"].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vects=CountVectorizer(binary=True)
```

```python
bags=vects.fit_transform(df["cleaned"])
print("vocabulary:", vects.get_feature_names_out())
print("bow matrix:\n", bags.toarray())
```

```python
y=df["Liked"]
x=bags
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report


nb=BernoulliNB()
nb.fit(x_train,y_train)
y_pred=nb.predict(x_test)
```

```python
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision:{precision_score(y_test, y_pred):.4f}")
print(f"Recall:   {recall_score(y_test, y_pred):.4f}")
print(f"F1 Score: {f1_score(y_test, y_pred):.4f}")
print('confusion matrix:',confusion_matrix(y_test, y_pred))
print('Classification Report:',classification_report(y_test, y_pred))
```

```python
import pickle
with open("review.pkl", "wb") as file:
    pickle.dump(nb, file)
```

```python
with open("review.pkl", "rb") as file:
    loaded_model = pickle.load(file)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle

# Fit the vectorizer to your training data
vect = TfidfVectorizer()
vect.fit(df['cleaned']) 

# Save the FITTED vectorizer
pickle.dump(vect, open('vectorize.pkl', 'wb'))
```
